```python
# Library Management System 
# Step 1
library = {
    "Python Basics": {"author": "Test", "available": True},
    "Data Science": {"author": "Test1", "available": False, "issued_to": "John Doe"}
}

credentials = {"admin": "", "librarian": ""}

# Display all books
def display_books():
    print("\nList of Books:")
    for title, details in library.items():
        status = "Available" if details["available"] else f"Issued to {details.get('issued_to', 'Unknown')}"
        print(f"Title: {title}, Author: {details['author']}, Status: {status}")

# Step 2 - Adding Books
def add_book():
    title = input("Enter the book title: ")
    author = input("Enter the author: ")
    if title in library:
        print(f"The book '{title}' already exists.")
    else:
        library[title] = {"author": author, "available": True}
        print(f"The book '{title}' by {author} has been added.")

# Step 3- Issuing Books
def issue_book():
    print("\nAvailable Books:")
    available_books = [title for title, details in library.items() if details["available"]]
    if not available_books:
        print("No books are currently available for issuing.")
        return
    for idx, book in enumerate(available_books, start=1):
        print(f"{idx}. {book}")
    try:
        choice = int(input("Enter the number of the book to issue: "))
        if 1 <= choice <= len(available_books):
            title = available_books[choice - 1]
            person = input("Enter the name of the person issuing the book: ")
            library[title]["available"] = False
            library[title]["issued_to"] = person
            print(f"The book '{title}' has been issued to {person}.")
        else:
            print("Invalid choice. Please select a valid number.")
    except ValueError:
        print("Invalid input. Please enter a number.")

# Step 4- Returning Books
def return_book():
    print("\nIssued Books:")
    issued_books = [title for title, details in library.items() if not details["available"]]
    if not issued_books:
        print("No books are currently issued.")
        return # Stop the function if no books are issued
    for idx, book in enumerate(issued_books, start=1):
        print(f"{idx}. {book}")
    try:
        choice = int(input("Enter the number of the book to return: "))
        if 1 <= choice <= len(issued_books):
            title = issued_books[choice - 1]
            library[title]["available"] = True
            library[title].pop("issued_to", None)
            print(f"The book '{title}' has been returned and is now available.")
        else:
            print("Invalid choice. Please select a valid number.")
    except ValueError:
        print("Invalid input. Please enter a number.")

# Step 5 - Login System
def login():
    print("\nLogin to the Library Management System")
    for _ in range(3):  
        username = input("Username: ")
        password = input("Password: ")
        if credentials.get(username) == password:
            print("Login successful!")
            return True
        else:
            print("Invalid credentials. Try again.")
    print("Too many failed attempts. Exiting the system.")
    return False

# Step 6 - View Issued Books
def view_issued_books():
    print("\nIssued Books:")
    issued_books = {title: details for title, details in library.items() if not details["available"]}
    if not issued_books:
        print("No books are currently issued.")
    else:
        for title, details in issued_books.items():
            print(f"Title: {title}, Issued to: {details.get('issued_to', 'Unknown')}")

# Step 7 - Main Menu
def main_menu():
    if not login():
        return  
    while True:
        print("\nLibrary Management System")
        print("1. View Books")
        print("2. Add a Book")
        print("3. Issue a Book")
        print("4. Return a Book")
        print("5. View Issued Books")
        print("6. Exit")
        choice = input("Enter your choice: ")
        if choice == "1":
            display_books()
        elif choice == "2":
            add_book()
        elif choice == "3":
            issue_book()
        elif choice == "4":
            return_book()
        elif choice == "5":
            view_issued_books()
        elif choice == "6":
            print("Exiting the system. Goodbye!")
            break
        else:
            print("Invalid choice. Please try again.")

# Run the program
main_menu()
```
